````python
import math
a = []  # заводим пустой список
n = int(input('Количество элементов в списке n='))  # считываем количество элемент в списке
for i in range(n):  
    new_element = int(input())  # считываем очередной элемент
    a.append(new_element)  # добавляем его в список
    # последние две строки можно было заменить одной:
    # a.append(int(input()))
print('Сформированный список ', a)
print("максимальный элемент списка ", max(a))
s= max(a)
print("Минимальный элемент списка ", min(a))
print(s)
````

Возведение матрицы в степень

````python
import numpy as np

from numpy import linalg as ln

f = np.array([[-1, -1, 0], [-1, 0, 1], [-2,-1,1]])

n = int(input('Степень для возведения матрицы n = '))

for i in range(1, n):
  print(i, ': ')
  print(ln.matrix_power(f, i))

````

````python
from numpy import linalg as ln
f = np.array([[1, -1, 3], [-1, -2, 1], [-2,-1, 3]])

n = int(input('Степень для возведения матрицы n = '))

for i in range(1, n):
  print(i, ': ')
  print(ln.matrix_power(f, i))
````

````python
print('a = ')
print(np.array([[1, -1, 2], [-1, 2, 0], [-2, 1, -1]]))
a = np.array([[1, -1, 2], [-1, 2, 0], [-2, 1, -1]])
print('at = ')
at  = a.transpose()
print(at)
print('b = ')
b = np.array([[-1, 2, 0], [-2, 1, 0], [-1, 0, -1]])
print(b)
bt = b.transpose()
print('bt = ')
print(bt)
#Property transpose matrix:1
c = a+b
print('a+b=')
print(a+b)
a_b_t = (a+b).transpose()
print('(a+b)T = ')
print(a_b_t)
print('at + bt = ')
at_bt = at + bt
print(at_bt)
a_b_t != at_bt
print(a_b_t != at_bt)
````
